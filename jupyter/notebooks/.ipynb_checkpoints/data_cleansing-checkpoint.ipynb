```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine, text
```

```python
engine = create_engine(
            'postgresql+psycopg2://admin:admin@de_psql:5432/de_psql',
            connect_args={'options': '-csearch_path=dwh'}
        )
```

```python
query = '''SELECT 
    c.symbol,
    t.full_date AS date,
    f.open_price,
    f.high_price,
    f.low_price,
    f.close_price,
    f.volume
FROM dwh.fact_history_stock f
JOIN dwh.dim_time t ON f.date_key = t.date_key
JOIN dwh.dim_company_informations c ON f.company_key = c.company_key;'''
df = pd.read_sql(query, engine)
df
```

```python
def get_technical_indicators(dataset):
    # -------------------- SMA --------------------
    dataset['sma7'] = dataset['close_price'].rolling(window=7).mean()
    dataset['sma21'] = dataset['close_price'].rolling(window=21).mean()
    # -------------------- MACD --------------------
    dataset['ema26'] = dataset['close_price'].ewm(span=26, adjust=False).mean()
    dataset['ema12'] = dataset['close_price'].ewm(span=12, adjust=False).mean()
    dataset['MACD'] = dataset['ema12'] - dataset['ema26']
    # -------------------- Bollinger Bands --------------------
    dataset['20sd'] = dataset['close_price'].rolling(window=21).std()
    dataset['upper_band'] = dataset['sma21'] + (dataset['20sd']*2)
    dataset['lower_band'] = dataset['sma21'] - (dataset['20sd']*2)
    # -------------------- RSI --------------------
    delta = dataset['close_price'].diff()
    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)
    roll_up = gain.rolling(window=14).mean()
    roll_down = loss.rolling(window=14).mean()
    RS = roll_up / roll_down
    dataset['RSI'] = 100.0 - (100.0 / (1.0 + RS))
    # -------------------- ATR --------------------
    dataset['H-L'] = dataset['high_price'] - dataset['low_price']
    dataset['H-PC'] = abs(dataset['high_price'] - dataset['close_price'].shift(1))
    dataset['L-PC'] = abs(dataset['low_price'] - dataset['close_price'].shift(1))
    dataset['TR'] = dataset[['H-L', 'H-PC', 'L-PC']].max(axis=1)
    dataset['ATR'] = dataset['TR'].rolling(window=14).mean()
    # -------------------- Return % --------------------
    dataset['return_pct'] = dataset['close_price'].pct_change() * 100
    # -------------------- Volatility 14 --------------------
    dataset['volatility_14'] = dataset['return_pct'].rolling(window=14).std()
    
    
    # -------------------- Drop temporary columns --------------------
    dataset.drop(columns=['20sd', 'H-L', 'H-PC', 'L-PC', 'TR'], inplace=True)
    
    return dataset
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
print(df['date'].dtype)
numeric_cols = ['open_price', 'high_price', 'low_price', 'close_price', 'volume']
df = df.groupby('symbol', group_keys=False).apply(get_technical_indicators)
# Xóa các dòng nếu cột 'open_price' hoặc cột kết quả hàm def bị NaN
df_cleaned = df.dropna(subset=['sma7', 'sma21']).reset_index(drop=True)
df_cleaned
```
